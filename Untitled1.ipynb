```python
import random

def read_file(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        words = [word.strip() for word in file.readlines()]
    return words

def word_progress(word, guessed_letters):
    progress = []
    for letter in word:
        if letter in guessed_letters:
            progress.append(letter)
        else:
            progress.append('_')
    return ' '.join(progress)

def user_guess(alphabet=set('абвгдеёжзийклмнопрстуфхцчшщъыьэюя')):
    text = input("Введите одну букву: ")
    if len(text) != 1:
        print("нужно одну букву, много букв сложно")
        return
    if text.lower() not in alphabet:
        print("Нужна одна буква из русского алфавита, не используй эльфийский или кракозябры")
        return
    return text.lower()

def play_game():
    words = read_file('input.txt')
    word = random.choice(words).lower()
    guessed_letters = []
    attempts = len(word) * 2

    print(f"Загадано слово из {len(word)} букв.")

    while attempts > 0:
        print(word_progress(word, guessed_letters))
        print(f"У вас осталось {attempts} попыток.")

        guess = user_guess()
        if not guess:
            continue

        if guess in guessed_letters:
            print("Вы уже пытались эту букву.")
            continue

        guessed_letters.append(guess)
        if guess not in word:
            attempts -= 1
            print(f"Неверная буква. Осталось {attempts} попыток.")
        else:
            print("Верно!")

        if '_' not in word_progress(word, guessed_letters):
            print(f"Поздравляю! Вы угадали слово '{word}'.")
            return

    print(f"Игра окончена. Загаданное слово было '{word}'.")

if __name__ == "__main__":
    play_game()
```

```python
import random

def is_valid_input(user_input):
    if user_input.isdigit():
        return True
    else:
        return False

def play_game():
    print("Я загадываю число от 1 до 100. Угадать его!")

    number_to_guess = random.randint(1, 100)
    attempts = 0

    while True:
        user_input = input("Введите ваше предположение: ")

        if not is_valid_input(user_input):
            print("Пожалуйста, вводите только целые числа.")
            continue

        guess = int(user_input)
        attempts += 1
        if guess == number_to_guess:
            print(f"Поздравляю! Вы угадали число {number_to_guess} за {attempts} попыток.")
            break
        elif random.randint(1, 12-attempts) == 1:
          print(f"Слишком много думаешь, ты проиграл. Правильный ответ {number_to_guess}")
          break
        elif random.randint(1, 4) == 1:
            print("Сегодня мне лень, думай сам!")
        elif guess < number_to_guess:
            print("маленькое число")
        elif guess > number_to_guess:
            print("большое число")


if __name__ == "__main__":
    play_game()
```

```python
import random

def play_game(num_questions=5):
    correct_answers = 0
    print(f"Вы должны ответить на {num_questions} вопросов.")

    for _ in range(num_questions):
        num1 = random.randint(1, 1000)
        num2 = random.randint(1, 1000)
        correct_answer = num1 * num2

        user_input = input(f"\nСколько будет {num1} умножить на {num2}? ")

        if user_input.isdigit():
            user_answer = int(user_input)

            if user_answer == correct_answer:
                print("Правильно!")
                correct_answers += 1
            else:
                print(f"Неправильно! Правильный ответ: {correct_answer}. Вы проиграли.")
                return
        else:
            print("Пожалуйста, введите корректное целое число. Вы проиграли.")
            exit()

    print(f"Игра закончена! Вы ответили правильно на {correct_answers} из {num_questions} вопросов.")

if __name__ == "__main__":
    play_game()
```

```python
import pandas as pd

df = pd.read_csv('data.csv')

print("Начальные цены:")
print(df['Price'].tail(10))

last_ten_rows = df.tail(10)
last_ten_rows = last_ten_rows.drop(columns=['Price'])

combined_df = pd.concat([df, last_ten_rows], ignore_index=True)
average_price = df['Price'].mean()
combined_df['Price'].fillna(average_price, inplace=True)

print("\nЦены после преобразования:")
print(combined_df['Price'])
```
